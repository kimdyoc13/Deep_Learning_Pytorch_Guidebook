### 소프트 맥스 함수

- 입력 값이 0 ~ 1 에서 출력
- 출력 값의 총합 이 1 이다.
- 출력 노드의 활성화 함수로 사용


image.png

렐루 함수와 소프트 맥스 함수의 파이토치 구현

```python
import torch
class Net(torch. nn.Module):
  def __init__(self, n_feature, n_hidden, n_output):
    super(Net, self).__init__()
    self.hidden = torch.nn.Linear(n_feature, n_hidden) # 은닉층
    self.relu = torch.nn.ReLu(inplace = True)
    self.out = torch.nn.Linear(n_hidden, n_output) # 출력층
    self.softmax = torch.nn.Softmax(dim = n_output)

  def forword(self, x):
    x = self.hidden(x)
    x = self.relu(x) # 은닉층을 위한 렐루 활성화 함수
    x = self.out(x)
    x = self.softmax(x) # 출력층을 위한 소프트맥스 활성화 함수
    return x
```

# 4.3 딥러닝 알고리즘
딥러닝은 심층신경망을 사용한다.

심층 신경망은

- 합성곱 신경망(CNN)
- 제한된 볼츠만 머신(RBM)
- 심층 신뢰 신경망(DBN)

등으로 분류 된다

# 4.3.1 심층 신경망
- 입력층과 출력층 사이에 다수의 은닉층을 포함한 인공 신경망
- 다수의 은닉층 덕분에 비선형 분류가 가능해짐

- 학습을 위한 연산량이 많고, 기울기 소멸 문제 등이 있음
  - 이를 해결하기 위해
    - 드롭아웃
    - 렐루 함수
    - 배치 정규화  진행

image.png

심층 신경망의 사진

# 4.3.2 합성곱 신경망
- 합성곱층 과 풀링층을 포함
- 이미지 처리, 객체 탐지 등에 좋은 성능을 내는 인공 신경망 알고리즘

image.png

예시
- LeNet-5
- AlexNet
- VGG
- GoogleNet
- ResNet

특징
- 각 층의 입출력 형상을 유지
- 이미지의 공간 정보를 유지하면서 인접 이미지와 차이가 있는 특징을 효과적으로 인식
- 복수 필터로 이미지의 특징을 추출하고 학습
- 추출한 이미지의 특징을 모으고 강화하는 폴링층 존재
- 필터를 공유 파라미터로 사용하기 때문에 일반 신경망과 비교하여 학습 파라미터가 매우 적음

# 4.3.3 순환 신경망
- 시계열 데이터(음악, 영상) 같은 시간 흐름에 따라 변화하는 데이터를 학습하기 위한 인공 신경망
- '순환' 이란 자기 자신을 참조한다는 것으로, 현재 결과가 이전 결과에 연관이 있음을 의미


image.png

특징
- 시간성을 가진 데이터가 많음
- 시간성 정보를 이용하여 데이터의 특징을 다룸
- 시간에 따라 내용이 변하므로 데이터는 동적이고, 길이가 가변적임
- 매우 긴 데이터를 처리하는 연구가 활발히 진행중
- 보통 Gradient Vanishing 문제가 자주 발현됨    
그래서 LSTM(Long Short Term Memory) 가 사용
- 자연어 처리와 궁합이 맞음
  - 언어 모델링
  - 텍스트 생성
  - 자동 번역
  - 음성 인식
  - 이미지 캡션 생성

# 4.3.4 제한된 볼츠만 머신
볼츠만 머신
- 가시층과 은늑층으로 구성된 모델

- 가시층과 은닉층이 제한된 것이 제한된 볼츠만 머신(Restricted Boltzmann Machine, RBM)

image.png

- 차원 감소, 분류, 선형 회귀 분석, 협업 필터링, 특성 값 학습, 주제 모델링 에 사용
- 기울기 소멸 문제를 해결하기 위해 사전 학습 용도로 활용 가능
- 심층 신뢰 신경망(DBN)의 요소로 활용

제일 많이 사용되는 알고리즘은 CNN 과 RNN 이다.


# 4.3.5 심층 신뢰 신경망

- 입력층과 은닉층으로 구성된 제한된 볼츠만 머신을 블록처럼 여러층으로 쌓은 형태로
연결된 신경망
- 사전 훈련된 제한된 볼츠만 머신을 층층이 쌓아 올린 구조
- 비지도 학습 가능


학습 절차
1. 가시층과 은닉층 1에 제한된 볼츠만 머신을 사전 훈련합니다.
2. 첫 번째 층 입력 데이터와 파라미터를 고정하여 두 번째 층 제안된 볼츠만 머신을 사전 훈련 함
3. 원하는 층 개수 만큼 제한된 볼츠만 머신을 쌓아 올려 전체 DBN 을 완성함

- 순차적으로 심층 신뢰 신경망을 학습시켜 사면서 계층적 구조를 생성
- 비지도 학습으로 학습
